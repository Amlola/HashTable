```python
def draw_bar(filename):
  df = pd.read_csv(filename, header=None)
  func_name = filename.split('_')[1].split('.')[0]
  df = df.rename(columns={0: 'list_index', 1: 'elements_number'})
  #min_y = 0
  #max_y = df['elements_number'].max()
  #min_x = 0
  #for i in range(df.shape[0]):
  #  if df.iloc[i]['elements_number'] != 0:
  #    min_x = i
  #    break
  #max_x = 0
  #for i in range(df.shape[0] - 1, -1, -1):
  #  if df.iloc[i]['elements_number'] != 0:
  #    max_x = i
  #    break
  func_name = filename.split('_')[1].split('.')[0]
  fig = px.bar(df, x='list_index', y='elements_number', title=func_name + " stats")
  #fig.update_xaxes(range=[max(min_x - 1, 0), max_x + 1])
  #fig.update_yaxes(range=[min_y, max_y])
  fig.update_traces(dict(marker_line_width=0))
  fig.show()
  name = filename.split('.')[0]
  fig.write_image(name + '.png')
```

```python
draw_bar("hashfunc_[0].csv")
```
